```python
classes=data["Disease"].nunique()
print(classes)
data["Disease"].unique()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder=LabelEncoder()

data["Disease"]=encoder.fit_transform(data["Disease"])
```

```python
x_data=data.drop("Disease",axis=1)
y_data=data.iloc[:,-1:]
print(x_data.shape)
print(y_data.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,test_size=0.2,random_state=42)
```

```python
y_train=np.asarray(y_train)
y_test=np.asarray(y_test)
x_train=np.asarray(x_train)
x_test=np.asarray(x_test)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#reshaping
x_train=x_train.reshape(-1,6,4)
x_test=x_test.reshape(-1,6,4)

from tensorflow.keras.layers import Dense,Flatten,MaxPooling2D,Conv2D,LSTM,Dropout,Bidirectional,Reshape,GlobalMaxPooling1D
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping

model=Sequential([
    Bidirectional(LSTM(128,activation="tanh",input_shape=(6,4),return_sequences=True)),
    GlobalMaxPooling1D(),
    Bidirectional(LSTM(64,activation="tanh",return_sequences=True)),
    GlobalMaxPooling1D(),
    Bidirectional(LSTM(64,activation="tanh",return_sequences=True)),
    Bidirectional(LSTM(32,activation="tanh")),
    Reshape((-1,16,2,2)),
    Conv2D(64,(1,1),activation="relu",padding="same",input_shape=(6,4,1)),
    MaxPooling2D((1,1)),
    Conv2D(32,(1,1),activation="relu",padding="same"),
    MaxPooling2D((1,1)),
    Flatten(),
    Dense(64,activation="relu"),
    Dense(16,activation="relu"),
    Dense(classes,activation="sigmoid")
    
])
```

```python
#reshaping of the data
x_train = x_train.reshape(-1,6, 4, 1)
x_test = x_test.reshape(-1,6, 4, 1)

from tensorflow.keras.layers import Dense,Flatten,MaxPooling2D,Conv2D,LSTM,Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping


model=Sequential([
    Conv2D(64,(1,1),activation="relu",padding="same",input_shape=(6,4,1)),
    MaxPooling2D((1,1)),
    Dropout(0.2),
    Conv2D(64,(1,1),activation="relu",padding="same"),
    MaxPooling2D((1,1)),
    Dropout(0.2),
    Conv2D(64,(1,1),activation="relu",padding="same"),
    MaxPooling2D((1,1)),
    Dropout(0.2),
    Flatten(),
    Dense(64,activation="relu"),
    Dense(16,activation="relu"),
    Dense(classes,activation="softmax")
])
```

```python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])
early_stopping=EarlyStopping(monitor="val_loss",patience=10,restore_best_weights=True)
```

```python
history=model.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=15, callbacks=[early_stopping])
```

```python
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.plot(history.history["loss"], label="Training Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
